```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from textblob import TextBlob  #sentiment analysis

# Extracting key themes
key_themes = ['quality', 'delivery', 'customer service', 'price', 'menu', 'cleanliness', 'online', 'staff', 'time', 'money', 'package']

# Categorizing reviews based on key themes
for theme in key_themes:
    pizza_reviews[theme] = pizza_reviews['Review'].str.contains(theme, case=False)

# Sentiment Analysis
def analyze_sentiment(review):
    analysis = TextBlob(review)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'

pizza_reviews['Sentiment'] = pizza_reviews['Review'].apply(analyze_sentiment)
```

```python
# Load the CSV file into a Pandas DataFrame
df_subset = pd.read_csv('sentiments_with_dates_updated.csv')
```

```python
df_subset['Date'] = pd.to_datetime(df_subset['Date'])

# Add a new column for the month
df_subset['Month'] = df_subset['Date'].dt.to_period('M')

# Filter positive and negative sentiments
positive_reviews = df_subset[df_subset['Sentiment'] == 'positive']
negative_reviews = df_subset[df_subset['Sentiment'] == 'negative']

# Group by month and count the number of positive and negative reviews
positive_counts = positive_reviews.groupby('Month').size()
negative_counts = negative_reviews.groupby('Month').size()

# Create a DataFrame with the counts
review_counts = pd.DataFrame({
    'Positive': positive_counts,
    'Negative': negative_counts
})

# Calculate MoM increase
review_counts['MoM_Positive_Increase'] = review_counts['Positive'].pct_change()
review_counts['MoM_Negative_Increase'] = review_counts['Negative'].pct_change()

# Remove rows where MoM_Positive_Increase and MoM_Negative_Increase are NaN
review_counts = review_counts.dropna(subset=['MoM_Positive_Increase', 'MoM_Negative_Increase'])

# Round to two decimals and include the percentage sign
review_counts['MoM_Positive_Increase'] = review_counts['MoM_Positive_Increase'].round(2).astype(str) + '%'
review_counts['MoM_Negative_Increase'] = review_counts['MoM_Negative_Increase'].round(2).astype(str) + '%'

# Display the result
print(review_counts)
```

```python
review_counts.to_csv('MoM_review_increase.csv', index=True)
```

```python
# Frequency calculation for each value in key_themes columns
frequency_data = {}

for theme in key_themes:
    theme_counts = pizza_reviews[theme].value_counts()
    frequency_data[theme] = theme_counts

# Display the frequency data
for theme, counts in frequency_data.items():
    print(f"\nFrequency counts for {theme}:")
    print(counts)
```

```python
# Create a new DataFrame to store the count of True values for each theme and sentiment
summary_data = pd.DataFrame(index=['positive', 'neutral', 'negative'], columns=key_themes + ['Sentiment Count'])

# Loop through sentiments and key_themes to populate the summary_data DataFrame
for sentiment in ['positive', 'neutral', 'negative']:
    for theme in key_themes:
        true_count = pizza_reviews[pizza_reviews['Sentiment'] == sentiment][theme].sum()
        summary_data.at[sentiment, theme] = true_count
    
    # Add the count of values in 'Sentiment' column
    sentiment_count = len(pizza_reviews[pizza_reviews['Sentiment'] == sentiment])
    summary_data.at[sentiment, 'Sentiment Count'] = sentiment_count

display(summary_data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords  # Make sure you have nltk installed: pip install nltk

positive_reviews = pizza_reviews[pizza_reviews['Sentiment'] == 'positive']['Review']
neutral_reviews = pizza_reviews[pizza_reviews['Sentiment'] == 'neutral']['Review']
negative_reviews = pizza_reviews[pizza_reviews['Sentiment'] == 'negative']['Review']

# Combine the reviews into a list for each sentiment
positive_text = ' '.join(positive_reviews)
neutral_text = ' '.join(neutral_reviews)
negative_text = ' '.join(negative_reviews)

# Create a CountVectorizer
vectorizer = CountVectorizer(stop_words=stopwords.words('english'), max_features=20)  # Adjust max_features as needed

# Fit and transform the positive, neutral, and negative text data
positive_word_matrix = vectorizer.fit_transform([positive_text])
neutral_word_matrix = vectorizer.fit_transform([neutral_text])
negative_word_matrix = vectorizer.fit_transform([negative_text])

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Convert word matrices to DataFrames for better visualization
positive_word_df = pd.DataFrame(positive_word_matrix.toarray(), columns=feature_names)
neutral_word_df = pd.DataFrame(neutral_word_matrix.toarray(), columns=feature_names)
negative_word_df = pd.DataFrame(negative_word_matrix.toarray(), columns=feature_names)

# Display the most frequent words in positive, neutral, and negative reviews
print("Most Frequent Words in Positive Reviews:")
print(positive_word_df.sum().sort_values(ascending=False))

print("\nMost Frequent Words in Neutral Reviews:")
print(neutral_word_df.sum().sort_values(ascending=False))

print("\nMost Frequent Words in Negative Reviews:")
print(negative_word_df.sum().sort_values(ascending=False))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords  # Make sure you have nltk installed: pip install nltk

# Assuming you have a list of positive, neutral, and negative reviews
positive_reviews = pizza_reviews[pizza_reviews['Sentiment'] == 'positive']['Review']
neutral_reviews = pizza_reviews[pizza_reviews['Sentiment'] == 'neutral']['Review']
negative_reviews = pizza_reviews[pizza_reviews['Sentiment'] == 'negative']['Review']

# Combine the reviews into a list for each sentiment
positive_text = ' '.join(positive_reviews)
neutral_text = ' '.join(neutral_reviews)
negative_text = ' '.join(negative_reviews)

# Create a CountVectorizer
vectorizer = CountVectorizer(stop_words=stopwords.words('english'))

# Fit and transform the positive text data
positive_word_matrix = vectorizer.fit_transform([positive_text])
positive_feature_names = vectorizer.get_feature_names_out()

# Fit and transform the neutral and negative text data using the same feature names
neutral_word_matrix = vectorizer.transform([neutral_text])
negative_word_matrix = vectorizer.transform([negative_text])

# Convert word matrices to DataFrames using the shared feature names
positive_word_df = pd.DataFrame(positive_word_matrix.toarray(), columns=positive_feature_names)
neutral_word_df = pd.DataFrame(neutral_word_matrix.toarray(), columns=positive_feature_names)
negative_word_df = pd.DataFrame(negative_word_matrix.toarray(), columns=positive_feature_names)

# Get words with a frequency over 5 in any review
common_words_over_5 = set(positive_word_df.columns[positive_word_df.sum() > 5]) | \
                     set(neutral_word_df.columns[neutral_word_df.sum() > 5]) | \
                     set(negative_word_df.columns[negative_word_df.sum() > 5])

# Display the common words with frequency over 5
print("Words with Frequency Over 5 in Any Review:")
print(common_words_over_5)
```

```python
import spacy
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import numpy as np

# Create a DataFrame with common_words_over_5
common_words_df = pd.DataFrame(common_words_over_5, columns=['Word'])

# Use spaCy for word embeddings
nlp = spacy.load("en_core_web_md")

# Get word vectors for common words
common_words_df['Vector'] = common_words_df['Word'].apply(lambda x: nlp(x).vector)

# Filter out zero vectors
common_words_df = common_words_df[common_words_df['Vector'].apply(lambda x: np.any(x != 0))]

# Perform hierarchical clustering
if len(common_words_df) > 0:
    word_vectors = list(common_words_df['Vector'])
    clustering = AgglomerativeClustering(n_clusters=5, affinity='cosine', linkage='complete')
    common_words_df['Cluster'] = clustering.fit_predict(word_vectors)

    # Display the resulting clusters
    print(common_words_df[['Word', 'Cluster']])
else:
    print("No valid word vectors for clustering.")
```

```python
reviews = pizza_reviews['Review'].tolist()
sentiments = pizza_reviews['Sentiment'].tolist()

# Clusters from your previous results
cluster1 = common_words_df[common_words_df['Cluster'] == 0]["Word"].values
cluster2 = common_words_df[common_words_df['Cluster'] == 1]["Word"].values
cluster3 = common_words_df[common_words_df['Cluster'] == 2]["Word"].values
cluster5 = common_words_df[common_words_df['Cluster'] == 4]["Word"].values

clusters = [cluster1, cluster2, cluster3, cluster5]

# Create a DataFrame to store the counts
cluster_counts_df = pd.DataFrame(index=['positive', 'negative', 'neutral'], columns=['Cluster1', 'Cluster2', 'Cluster3', 'Cluster5'])

# Minimum occurrences required in a review
min_occurrences = 10

# Iterate through clusters
for i, cluster in enumerate(clusters):
    # Iterate through sentiments
    for sentiment in ['positive', 'negative', 'neutral']:
        # Count the occurrences of words in the cluster in reviews of the given sentiment
        count = sum(sum(word in review for word in cluster) >= min_occurrences and sentiment == s for review, s in zip(reviews, sentiments))
        # Update the DataFrame
        if i != 3:
            cluster_counts_df.at[sentiment, f'Cluster{i + 1}'] = count
        if i == 3:
            cluster_counts_df.at[sentiment, f'Cluster{i + 2}'] = count

# Display the final DataFrame
print(cluster_counts_df)
```
